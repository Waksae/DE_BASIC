```python
import requests
import pandas as pd

API_KEY = '566577796674757433315742734259'
TYPE = 'json'
SERVICE = 'StationDstncReqreTimeHm'
START_INDEX = 1
END_INDEX = 102

def get_info(API_KEY):
    url = f'http://openapi.seoul.go.kr:8088/{API_KEY}/{TYPE}/{SERVICE}/{START_INDEX}/{END_INDEX}/'
    try:
        res = requests.get(url, timeout = 5)
        print(f'응답코드 : {res.status_code}')
        # print(f'응답본문 : {res.text}')
        if res.status_code == 200:
            data = res.json()
            return data['StationDstncReqreTimeHm']['row']
        else:
            print('응답 없음', res.text)
    except Exception as e:
        print(f"조회 오류 : {e}")
    return "조회 불가"

raw_data = get_info(API_KEY)
raw_data

cleaned_data = []
for i in raw_data :
    cleaned_data.append({
        "호선" : f"{i['SBWY_ROUT_LN']}호선",
        "역명" : i['SBWY_STNS_NM'],
        "이동시간" : i['HM'],
        "거리" : f"{i['DIST_KM']}km",
        "누적거리" : f"{i['ACML_DIST']}km",
})
df = pd.DataFrame(cleaned_data)

df.to_csv("역간 거리 정보.csv", index=False, encoding="utf-8-sig")
print("CSV 저장 완료: 역간 거리 정보.csv")
```
